```python
bat = % sql select * from batting where yearid>=2010
bat_df = bat.DataFrame()
bat_df.tail(10)
```

```python
import pandas as pd
bat_g = bat_df.groupby("playerID")
```

```python
bat_g_combined=pd.DataFrame()
bat_g_combined["H"] = bat_g['H'].sum()
bat_g_combined["HR"] = bat_g['HR'].sum()
bat_g_combined["2B"] = bat_g['2B'].sum()
bat_g_combined["3B"] = bat_g['3B'].sum()
bat_g_combined["1B"] = bat_g_combined['H']-bat_g_combined['2B']-bat_g_combined['3B']-bat_g_combined['HR']
bat_g_combined["AB"] = bat_g["AB"].sum()
bat_g_combined["SLP"] = (bat_g_combined['1B']+bat_g_combined['2B']*2+bat_g_combined['3B']*3+bat_g_combined['HR']*4) \
    / bat_g_combined["AB"]
```

```python
bat_g_combined = bat_g_combined.sort_values('SLP',ascending=False)
bat_g_combined[bat_g_combined["AB"] > 500]
```

```python
t = bat_g_combined['AB']>2000
t2 = bat_g_combined['H']>1000
bat_g_combined[t2 & t].sort_values('SLP',ascending=False)
```
